```python
# import the dotenv variables

from dotenv import load_dotenv

load_dotenv(override=True)

# import the openai library

from openai import OpenAI

# set the api key
import os
openai_api_key = os.getenv('OPENAI_API_KEY')

if openai_api_key:
    print(f"OpenAI API Key exists and begins {openai_api_key[:8]}")
else:
    print("OpenAI API Key not set - please head to the troubleshooting guide in the setup folder")

openai_client = OpenAI()

# create a list of messages in the familiar OpenAI format  
question = "I am an experienced leader in a large bank and I want to pick a business area that might be worth exploring for an Agentic AI opportunity that will be a market differeniator for us. I want you to think step by step and provide me with a list of business areas that I should consider." 
messages = [
    {"role": "user", "content": question}
]

print(messages)
```

```python

# call the OpenAI API
response = openai_client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages
)

answer = response.choices[0].message.content

print(answer)
```

```python
from IPython.display import Markdown, display

display(Markdown(question))
display(Markdown(answer))
```

```python
messages.append({"role": "user", "content": answer})

print(messages)
```

```python
question = "What is a painpoint in one of these business areas that are identified as having high potential for Agentic AI value and that has a high market differentiation potential?"

# messages = [
#     {"role": "user", "content": question}
# ]
messages.append({"role": "user", "content": question})

print(messages)

response = openai_client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages
)

answer = response.choices[0].message.content

display(Markdown(answer))

```

```python
messages.append({"role": "user", "content": answer})

question = "Propose an Agentic AI solution for that painpoint"

messages.append({"role": "user", "content": question})

print(messages)

response = openai_client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages
)

answer = response.choices[0].message.content

display(Markdown(answer))

```
